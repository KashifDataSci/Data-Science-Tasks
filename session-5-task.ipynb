###`Q1:` Join Tuples if similar initial element
While working with Python tuples, we can have a problem in which we need to perform concatenation of records from the similarity of initial element. This problem can have applications in data domains such as Data Science.

For eg.
```
Input  : test_list = [(5, 6), (5, 7), (5, 8), (6, 10), (7, 13)]
Output : [(5, 6, 7, 8), (6, 10), (7, 13)]
```

````python
# write your code here
text_list=[(5,6),(5,7),(5,8),(6,10),(7,13)]
first_unique_num=[]
for i in text_list:
    first_unique_num.append(i[0])
first_unique_num=list(set(first_unique_num))
result=[]
for i in first_unique_num:
    result.append([i])
    for j in text_list:
        if j[0]==i:
            result[-1].append(j[1])
print(list(map(tuple,result)))
    
````

###`Q2:` Multiply Adjacent elements (both side) and take sum of right and lest side multiplication result.


For eg.
```
The original tuple : (1, 5, 7, 8, 10)
Resultant tuple after multiplication :

(1*5, 1*5+5*7, 7*5 + 7*8, 8*7 + 8*10, 10*8) -> (5, 40, 91, 136, 80)

output-(5, 40, 91, 136, 80)
```

````python
# write your code here
t=(1, 5, 7, 8, 10)
result=[]
result.append(t[0]*t[1])
for i in range(1,len(t)-1):
    result.append(t[i-1]*t[i]+t[i]*t[i+1])
result.append(t[-1]*t[-2])    
print(result)
````

###`Q4`: Count no of tuples, list and set from a list
```
list1 = [{'hi', 'bye'},{'Geeks', 'forGeeks'},('a', 'b'),['hi', 'bye'],['a', 'b']]

```
`Output:`

```
List-2
Set-2
Tuples-1
```

````python
# write your code here
list1 = [{'hi', 'bye'},{'Geeks', 'forGeeks'},('a', 'b'),['hi', 'bye'],['a', 'b']]
no_of_tuples,no_of_list,no_of_sets=(0,0,0)
for i in list1:
    if type(i)==tuple:
        no_of_tuples=no_of_tuples+1
    elif type(i)==list:
        no_of_list=no_of_list+1
    elif type(i)==set:
        no_of_sets=no_of_sets+1
    else:
        pass
print("List_{}\n set-{}\nTuples-{}".format(no_of_list,no_of_sets,no_of_tuples))
````

###`Q1:` Write a program to find set of common elements in three lists using sets.
```
Input : ar1 = [1, 5, 10, 20, 40, 80]
        ar2 = [6, 7, 20, 80, 100]
        ar3 = [3, 4, 15, 20, 30, 70, 80, 120]

Output : [80, 20]
```

````python
# write your code here
ar1 = [1, 5, 10, 20, 40, 80]
ar2 = [6, 7, 20, 80, 100]
ar3 = [3, 4, 15, 20, 30, 70, 80, 120]
ar1=set(ar1)
ar2=set(ar2)
ar3=set(ar3)
print(list((ar1 & ar2) & ar3)[::-1])
````

### `Q4`: find union of n arrays.

**Example 1:**

Input:
```bash
[[1, 2, 2, 4, 3, 6],
 [5, 1, 3, 4],
 [9, 5, 7, 1],
 [2, 4, 1, 3]]
```

Output:

```bash
[1, 2, 3, 4, 5, 6, 7, 9]
```

````python
# write your code here
from functools import reduce
inputlist=[[1, 2, 2, 4, 3, 6],
 [5, 1, 3, 4],
 [9, 5, 7, 1],
 [2, 4, 1, 3]]
inputlist=list(map(set,inputlist))
result=list(reduce(lambda s1,s2:s1 | s2,inputlist))
print(result)
````

### `Q2`: Replace words from Dictionary. Given String, replace itâ€™s words from lookup dictionary.

**Example 1:**

Input:

```bash
test_str = 'CampusX best for DS students.'
repl_dict = {"best" : "is the best channel", "DS" : "Data-Science"}
```

Output:

```bash
CampusX is the best channel for Data-Science students.
```

**Example 2:**

Input:
```bash
test_str = 'CampusX best for DS students.'
repl_dict = {"good" : "is the best channel", "ds" : "Data-Science"}
```

Output:
```bash
CampusX best for DS students.
```

````python
# write your code here
test_str = 'CampusX best for DS students.'
repl_dict = {"best" : "is the best channel", "DS" : "Data-Science"}
result=[]
for word in test_str.split():
    if word in repl_dict:
        result.append(repl_dict[word])
    else:
        result.append(word)
    result.append(" ")
print(''.join(result))
````
