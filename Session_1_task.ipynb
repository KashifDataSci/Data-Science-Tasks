### Q4:- Write a program to find the euclidean distance between two coordinates.Take both the coordinates from the user as input.

```python
import math
x1=float(input("Enter X1"))
x2=float(input("Enter X2"))
y1=float(input("Enter y1"))
y2=float(input("Enter y2"))
eulidis=math.sqrt((x2-x1)**2-(y2-y1)**2)
print("The euclidean Distance of entered coordinates is ",eulidis)
```

### Q6:- Write a program that will tell the number of dogs and chicken are there when the user will provide the value of total heads and legs.
#### For example: Input: heads -> 4 legs -> 12
Output: dogs -> 2 chicken -> 2

```python
ttl_heads = int(input("Enter the total Heads: "))
ttl_legs = int(input("Enter the total Legs: "))

# Calculate the number of dogs using the derived formula
# Dogs have 4 legs, chickens have 2. 
# The number of dogs is half of the difference between total legs and two times the total heads.
Dogs = (ttl_legs - 2 * ttl_heads) / 2
print("The number of Dogs is:", int(Dogs))

# Calculate the number of chickens
Chick = ttl_heads - Dogs
print("The number of Chickens is:", int(Chick))
```
