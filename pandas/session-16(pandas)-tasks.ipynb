### `Q-5.`Write a function to change the data type of given a column or a Series. Function takes series and data type as input, returns the converted series.
```
series = pd.Series([1,2,'Python', 2.0, True, 100])
change to float type data
```
Note: Read about pd.to_numeric()

````python
# code here
series=pd.Series([1,2,'Python', 2.0, True, 100])
dtype=series.dtype
def change_datatype(series,dtype):
    return pd.to_numeric(series,errors='coerce').astype(dtype)
converted_series=change_datatype(series,float)
print(converted_series)
````

### `Q-9`
    i. Read `items.csv` making `item_name` as index.
    ii. Show no of nan values
    ii. Item price is given in $, so convert it to rupees without currency symbol.
    iii. Make data type of newly made series as float.
    iv. Fill nan with mean of the series


How csv file looks

```
item_name	item_price
Chips and Fresh Tomato Salsa	$2.39
Izze	$3.39
Nantucket Nectar	$3.39
Chips and Tomatillo-Green Chili Salsa	$2.39
Chicken Bowl	$16.98

```

````python
# code here
# i - read csv and make 'item_name' as index-col
df=pd.read_csv('Datasets_for_tasks\items.csv',index_col='item_name')
# ii - show number of nan values
df.isnull().sum()
# iii- convert item price from USD to PKR

df['item_price']=df['item_price'].str.replace('$',"",regex=False).astype(float)
df['item_price']=df['item_price']*280
df['item_price']

# iv- fill nan values with mean of the prices


df['item_price'].fillna(df['item_price'].mean,inplace=True)
df['item_price'].isna().sum()
````

### `Q-10`:
    i. Find mean price
    ii. Find 30th and 6th percentile value
    iii. Plot Histogram on price with bin size 50
    iv. No of items price lies between [1000 to 2000]



````python
# code here
# i- find mean price
# i - Convert item_price to float (you already did this)


# ii - Find mean price

# Remove any spaces and convert to float, coercing errors to NaN
df['item_price'] = pd.to_numeric(df['item_price'].astype(str).str.strip(), errors='coerce')

# Now calculate mean
mean_price = df['item_price'].mean()
print("Mean Price:", mean_price)

# ii- 30th and 6th percentile value
thirty_per=df['item_price'].quantile(0.30)
sixth_per=df['item_price'].quantile(0.06)
print(thirty_per)
print(sixth_per)
 # iii-  plot histogram on bin size of 50
df['item_price'].plot(bins=50,kind='hist')


# iv - Number of items with price between 1000 and 2000
count_items = df['item_price'].between(1000, 2000).sum()
print(count_items)


````
